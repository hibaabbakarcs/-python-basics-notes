```python
import math
from turtle import *

# معادلة X للقلب
def hearta(k):
    return 15 * math.sin(k) ** 3

# معادلة Y للقلب
def heartb(k):
    return (12 * math.cos(k)
            - 5 * math.cos(2*k)
            - 2 * math.cos(3*k)
            - math.cos(4*k))

hideturtle()
speed(10)
bgcolor("black")
color("pink")

for i in range(6000):
    goto(hearta(i) * 20, heartb(i) * 20)

    for j in range(5):
        color("#f73487")
    goto(0, 0)

done()
```

```python
#5)Objects & References
a = [1, 2]
b = a
#The variable does not store the value directly;
#it stores a reference (pointer) to the object.
```

```python
#6)Aliasing
a = [1, 2]
b = a

b[0] = 100
print(a)  # [100, 2]
```

```python
#7)Cloning Lists
#Create a standalone copy
a = [1, 2, 3]
b = a[:]      # نسخة

#or
b = list(a)
print(b)
```

```python
#8)Methods on Lists
#List-specific functions most comman
l=[1,6,3,3,5,6,2]
l.append(4)
l.remove(2)
l.sort()
```

```python
#9)Append vs Concatenate
#append() → يضيف عنصر واحد
l = [1,2]
l.append(3)   # [1,2,3]

#+ → دمج قائمتين
[1,2] + [3,4]   # [1,2,3,4]
```

```python
#10)Non-Mutating Methods (Strings)
s = "hello"
s.upper()   # HELLO

```

```python
#11)String format method
name = "Heba"
"Hello {}".format(name)

#best (f-string)
name = "Heba"
print(f"Hello {name}")
```

```python
#12)Accumulator with
#-List:
result = []
for x in [1,2,3]:
    result.append(x * 2)

#-String:
result = ""
for ch in "abc":
    result += ch.upper()
```
